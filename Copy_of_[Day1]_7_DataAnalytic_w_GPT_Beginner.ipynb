```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming it's already loaded into a DataFrame named games_df)
df = pd.read_csv('/content/[Dataset]Video_Games.csv')
# Exclude non-numeric columns for visualization
numeric_columns = games_df.select_dtypes(include=['float64', 'int64']).columns

# 1. Top-selling Games
top_games = games_df.sort_values(by='Global_Sales', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='Global_Sales', y='Name', data=top_games, palette='viridis')
plt.title('Top 10 Best-Selling Games')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Game')
plt.show()

# 2. Genre Analysis
plt.figure(figsize=(12, 6))
sns.countplot(x='Genre', data=games_df, order=games_df['Genre'].value_counts().index, palette='muted')
plt.title('Number of Games by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.show()

# 3. Platform Performance
plt.figure(figsize=(12, 6))
sns.boxplot(x='Platform', y='Global_Sales', data=games_df, palette='pastel')
plt.title('Global Sales by Platform')
plt.xlabel('Platform')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()

# 4. Regional Preferences (Example: North America Sales)
plt.figure(figsize=(12, 6))
sns.histplot(games_df['NA_Sales'], bins=20, kde=True, color='blue')
plt.title('Distribution of North America Sales')
plt.xlabel('North America Sales (in millions)')
plt.ylabel('Frequency')
plt.show()

# 5. Publisher Analysis
top_publishers = games_df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_publishers.values, y=top_publishers.index, palette='rocket')
plt.title('Top 10 Publishers by Global Sales')
plt.xlabel('Global Sales (in millions)')
plt.ylabel('Publisher')
plt.show()

# 6. Impact of Ratings (Example: Critic Score vs. Global Sales)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Critic_Score', y='Global_Sales', data=games_df, color='green')
plt.title('Critic Score vs. Global Sales')
plt.xlabel('Critic Score')
plt.ylabel('Global Sales (in millions)')
plt.show()

# 7. Longevity of Games (Example: Release Year vs. Global Sales)
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year_of_Release', y='Global_Sales', data=games_df, estimator=sum, ci=None, color='purple')
plt.title('Global Sales Over Years')
plt.xlabel('Year of Release')
plt.ylabel('Global Sales (in millions)')
plt.show()

# 8. Seasonal Trends (Example: Yearly Sales)
yearly_sales = games_df.groupby('Year_of_Release')['Global_Sales'].sum()
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, color='orange')
plt.title('Yearly Global Sales Trend')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()

# 9. Correlation Analysis (Example: Correlation Heatmap)
plt.figure(figsize=(10, 8))
sns.heatmap(games_df[numeric_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# 10. Emerging Trends (Example: Yearly Genre Trends)
plt.figure(figsize=(12, 6))
sns.countplot(x='Year_of_Release', hue='Genre', data=games_df)
plt.title('Genre Trends Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.legend(title='Genre', loc='upper right')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/data.csv')

# Basic setup for plots
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Grade Distribution
plt.subplot(3, 2, 1)
sns.histplot(data['final_grade'], bins=10, kde=True)
plt.title('Final Grade Distribution')

# Grade Progression
plt.subplot(3, 2, 2)
sns.lineplot(data=data[['grade_1', 'grade_2', 'final_grade']].mean(axis=0), marker='o')
plt.title('Average Grade Progression')
plt.xlabel('Grade Period')
plt.ylabel('Average Grade')

# Parental Education vs Final Grade
plt.subplot(3, 2, 3)
sns.boxplot(x='mother_education', y='final_grade', data=data)
plt.title('Mother\'s Education vs Final Grade')
plt.xticks(rotation=45)

plt.subplot(3, 2, 4)
sns.boxplot(x='father_education', y='final_grade', data=data)
plt.title('Father\'s Education vs Final Grade')
plt.xticks(rotation=45)

# Study Time vs Final Grade
plt.subplot(3, 2, 5)
sns.boxplot(x='study_time', y='final_grade', data=data)
plt.title('Study Time vs Final Grade')
plt.xticks(rotation=45)

# School Support vs Final Grade
plt.subplot(3, 2, 6)
sns.boxplot(x='school_support', y='final_grade', data=data)
plt.title('School Support vs Final Grade')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Create additional plots for other insights
plt.figure(figsize=(15, 10))

# Activities vs Final Grade
plt.subplot(3, 2, 1)
sns.boxplot(x='activities', y='final_grade', data=data)
plt.title('Activities vs Final Grade')

# Internet Access vs Final Grade
plt.subplot(3, 2, 2)
sns.boxplot(x='internet_access', y='final_grade', data=data)
plt.title('Internet Access vs Final Grade')

# Alcohol Consumption vs Final Grade
plt.subplot(3, 2, 3)
sns.boxplot(x='weekday_alcohol', y='final_grade', data=data)
plt.title('Weekday Alcohol Consumption vs Final Grade')

plt.subplot(3, 2, 4)
sns.boxplot(x='weekend_alcohol', y='final_grade', data=data)
plt.title('Weekend Alcohol Consumption vs Final Grade')

# Health vs Final Grade
plt.subplot(3, 2, 5)
sns.boxplot(x='health', y='final_grade', data=data)
plt.title('Health vs Final Grade')

# Absences vs Final Grade
plt.subplot(3, 2, 6)
sns.scatterplot(x='absences', y='final_grade', data=data)
plt.title('Absences vs Final Grade')

plt.tight_layout()
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
